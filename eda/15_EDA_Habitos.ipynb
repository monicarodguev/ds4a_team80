```python
# Load a file as a DataFrame and assign to df
df = pd.read_excel("Data/Habitos.xlsx")
```

```python
df.keys()
```

```python
col = np.char.lower(np.array(df.keys(), dtype='U'))
col[0] = 'id'
df.columns = col
```

```python
df.info()
```

```python
# Cambiar texto a minuscula.
a = df['tipo'].dtype

for col in df.columns:
    if df[col].dtype == a:
        df[col] = df[col].str.lower()
```

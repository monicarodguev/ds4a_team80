```python
# Librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Base de interés
ruta = 'Data/'

DatosBasicos = pd.read_excel(ruta +'Datos basicos.xlsx')
```

## 1. Acotando Variables

### Datos Basicos

```python
def letra_lower( df, columna ):
    a = df[columna].dtype
    for col in df.columns:
        if df[col].dtype == a:
            df[col] = df[col].str.lower()
    return df
```

```python
# rename columnas
DatosBasicos = DatosBasicos.rename(columns={'FE_NACIMIENTO' : 'fecha_n', 'AcompaÃ±ante' : 'acompanante'})

# columnas a lower
DatosBasicos.columns = DatosBasicos.columns.str.lower()
```

```python
# fecha de nacimiento 
DatosBasicos[ 'fecha_n' ] = pd.to_datetime(DatosBasicos["fecha_n"])
DatosBasicos['year_n'] = DatosBasicos[ 'fecha_n' ].apply( lambda x : x.year )
DatosBasicos['month_n'] = DatosBasicos[ 'fecha_n' ].apply( lambda x : x.month )

# drop a columnas
DatosBasicos.drop(columns=['edad', 'preferencia', 'acompanante','regimen','tipoafiliacion','relacionlaboral'], inplace= True)

DatosBasicos
```

```python
# exporta
DatosBasicos.to_csv('DatosBasicos.csv',sep='|')
```
